## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# your code here

def random_generater(bottom = 0, top = 100, n = 1000):
    """
    Generate n random numbers between bottom and top
    """
    bottom = min(bottom, top)
    top = max(bottom, top)
    n = n
    if n < 0:
        raise ValueError("n must be a positive integer")
    res = [random.uniform(bottom, top) for _ in range(n) ]
    return res
    
```

```python
def generate_normal_random_numbers(mu, sigma, n):
    n = n
    if n < 0:
        raise ValueError("n must be a positive integer")
    res = [random.gauss(mu, sigma) for _ in range(n)]
    return res
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
def generate_expon_random_numbers(mean, n):
    n = n
    if n < 0:
        raise ValueError("n must be a positive integer")
    res = [np.random.exponential(mean) for _ in range(n)]
    return res
```

```python
set5 = generate_expon_random_numbers(10, 10)
set6 = generate_expon_random_numbers(10, 100) 
```

```python
plt.subplot(1, 2, 1)
plt.hist(set5, bins=100, edgecolor='black', alpha=0.7)
plt.title("Exponential Distribution (n=10)")
plt.xlabel("Values")
plt.ylabel("Frequency")

# Histogram for the dataset of 100 numbers
plt.subplot(1, 2, 2)
plt.hist(set6, bins=100, edgecolor='blue', alpha=0.7)
plt.title("Exponential Distribution (n=100)")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# In summary, while both histograms represent an exponential distribution, the one based on 100 samples is a more reliable and smooth portrayal of the theoretical distribution compared to the more variable and less informative histogram of the 10-sample dataset.
```

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here

mean = 10  
lambda_val = 1 / mean 
time_limit = 15  

# Calculate the probability using the CDF of the exponential distribution
probability_less15 = 1 - math.exp(-lambda_val * time_limit)

```

```python
# your answer here
# Hint: This is same as saying P(x<15)
print("Probability that a customer will spend less than fifteen minutes:", probability_less15)
```

What is the probability that the customer will spend more than 15 minutes

```python
# your code here
probability_more15 = 1 - probability_less15
print("Probability that a customer will spend more than fifteen minutes:", probability_more15)
```

```python
x = np.linspace(0, 60, 400)
# Calculate the probability density function (PDF) values for each x
pdf = lambda_val * np.exp(-lambda_val * x)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'b-', label='PDF of Exponential Distribution (mean=10)')

# Shade the area where t > 15 minutes
x_shade = np.linspace(time_limit, 60, 200)
plt.fill_between(x_shade, lambda_val * np.exp(-lambda_val * x_shade), 
                 color='red', alpha=0.5, label='P(T > 15 minutes)')

# Add a vertical line at t = 15 minutes
plt.axvline(x=time_limit, color='black', linestyle='--', label='t = 15 minutes')

plt.xlabel('Time (minutes)')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution of Time Spent in Bank')
plt.legend()
plt.grid(True)
plt.show()

```
